```python

recursions_start = 1
recursion_end = 50
starting_points = -100
end_points = 100
step = 1

x_vals = []
y_vals = []

for starting_point in range(starting_points,end_points, step):
    for recursion in range(recursions_start,recursion_end, step):
        x = src.numerical.bumpy_reduction(starting_point, recursion)
        y = recursion
        x_vals.append(x)
        y_vals.append(y)


title = "Bumpy Reduction"
x_label = "Starting Point"
y_label = "Recursions"

src.plotting.plot_data(x_vals,y_vals, title, x_label, y_label)
```
